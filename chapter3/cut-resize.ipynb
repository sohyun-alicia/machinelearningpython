```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("test.png")
```

```python
# 이미지 자르기
im2 = img[150:450, 150:450]
```

```python
# 이미지 크기 변경하기
img = cv2.resize(im2, (400, 400))
```

```python
# 크기 변경한 이미지 저장하기
cv2.imwrite("cut-resize.png", im2)
```

```python
# 이미지 출력하기
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```
